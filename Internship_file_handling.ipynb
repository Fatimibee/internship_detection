## Now satrt merging both real and fake dataset

```python
real_internship=pd.read_csv('internshala_real_internship.csv')

fake_internship=pd.read_csv('Final_fraud_Internship.csv', index_col=0)

fake_internship.columns
```

```python
real_internship['Description'][1]
```

```python
fake_internship['Description'][98]
```

```python
# I need only title and desciption on internship 

real_internship1=pd.DataFrame({
    'Title':real_internship['Title'],
    'Description':real_internship['Description'],
    'IS_Scam':real_internship['IS_Scam']
})
real_internship1.head()
```

```python
# lets concat both data file(real and fake)

joined_df=pd.concat([fake_internship,real_internship1], ignore_index=True)
joined_df
```

```python
# Shufffling data
final_df = joined_df.sample(frac=1).reset_index(drop=True)
final_df
```

```python
# save final combined dataset
final_df.to_csv("Final_internship_dataset.csv",index=False)
```
