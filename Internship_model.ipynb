```python
import pandas as pd
```

```python
Accurate_df=pd.read_csv('Final_internship_dataset.csv')
Accurate_df
```

```python
# EDA report generate
from ydata_profiling import ProfileReport

report=ProfileReport(Accurate_df)
report.to_file(output_file='internship_report.html')
```

```python
#see duplicates
Accurate_df[(Accurate_df.duplicated()) ]
```

```python
#remove duplicates
Accurate_df.drop_duplicates(keep='first',inplace=True)
Accurate_df
```

```python
Accurate_df['IS_Scam'].value_counts(normalize=True)
#Imbalanced dataset
```

```python
#filling na value with empty string
Accurate_df['Title']=Accurate_df['Title'].fillna('')
Accurate_df['Description']=Accurate_df['Description'].fillna('')

X=Accurate_df.iloc[:,:-1]
Y=Accurate_df['IS_Scam']
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from imblearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from imblearn.over_sampling import RandomOverSampler
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,stratify=Y,random_state=45)
print(X_train.shape)
print(Y_train.shape)
```

```python
#for encoding
vector=TfidfVectorizer()

preprocessing=ColumnTransformer([
    ('titleencode',vector,'Title'),
    ('descriptionencode',vector,'Description')
],remainder='passthrough')
```

```python
#model  because datset is imbalanced
from sklearn.naive_bayes import ComplementNB
model=ComplementNB()
```

```python
pipe=Pipeline(
    [
    ('trf1', preprocessing),
    ('oversample', RandomOverSampler(random_state=42)),
    ('trf2', model)
    ]
)
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_predict=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
report=classification_report(Y_test,Y_predict)
print(report)
```

```python
score=accuracy_score(Y_test,Y_predict)
score
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(pipe, X, Y, cv=5, scoring='accuracy')
print("Average accuracy:", scores.mean())
```
